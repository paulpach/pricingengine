# Price Optimization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pymc3 as pm
from scipy.stats import expon
from tabulate import tabulate
from bayes_opt import BayesianOptimization
from bayes_opt import UtilityFunction
```

## Target function

```python
dist = [
    ("hat", expon(-3, 5)), 
    # ("shirt", expon(-2, 7)), 
    # ("pants", expon(-4, 5)), 
    # ("shoes", expon(-2, 3)), 
    # ("socks", expon(-2, 10)),
]

noise = 0.01
days = 100

fig = plt.figure()

ax = plt.subplot()
for (label, d) in dist:
    x = np.linspace(0, 30, 30)
    plt.plot(x, d.pdf(x), label=label)

ax.set_title('Winning price distribution')
plt.legend(loc="upper right")

plt.show()

def plotRevenue():
    fig = plt.figure()

    for (label,d) in dist:
        ax = plt.subplot()
        x = np.linspace(0, 30, 30)
        plt.plot(x, d.pdf(x) * x, label=label)
        
    ax.set_title('Revenue distribution')
    plt.legend(loc="upper right")
    plt.show()

plotRevenue()
```

## Optimizing

```python
data = np.array([[0 for i in dist]])
headers = [item for (item, d) in dist]

table = tabulate(data, headers, tablefmt="fancy_grid")
print(table)

optimizer = BayesianOptimization(
    f=None,
    pbounds={item : (0, 30) for (item, d) in dist},
    verbose=2,
    random_state=1,
)
utility = UtilityFunction(kind="ucb", kappa=2.5, xi=0.0)

    
def plotStep(point, observation, step):
    fig = plt.figure()

    for (label,d) in dist:
        ax = plt.subplot()
        x = np.linspace(0, 30, 30)
        p = plt.plot(x, d.pdf(x) * x, label=label)
        color = p[0].get_color()
        pointx = point[label]
        pointy = observation[label]
        plt.plot(pointx, pointy, "o", color=color)
    
    ax.set_title(f'Experiment on day {step}')
    plt.legend(loc="upper right")
    plt.show()

def measure(price, dist):
    #return dist.pdf(price) * price
    return np.random.normal(loc=dist.pdf(price) * price, scale=noise)

revenue_history = []

# optimization loop
for i in range(days):
    next_point = optimizer.suggest(utility)
    observation = {item: measure(next_point[item], d) for (item, d) in dist}
    revenue = sum(observation.values())
    revenue_history.append(revenue)
    m = optimizer.max
    plotStep(next_point, observation, i)
    print (f"revenue for the day {revenue}")
    print (f"estimated max {optimizer.max}")
    #print (next_point, observation, sum(observation.values()))
    optimizer.register(params=next_point, target=revenue)

    
fig = plt.figure()
ax = plt.subplot()
revx = np.arange(0, days)
 
plt.plot(revx, revenue_history, "bo", lw=2, label="revenue per day")
df = pd.DataFrame({'revenue': revenue_history})
df['trend'] = df['revenue'].ewm(span=6, adjust=False).mean()
plt.plot(revx, df['trend'], "r-", lw=2, label="trend")
plt.legend()
plt.show()
```
